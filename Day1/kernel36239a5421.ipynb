```python
data = pd.read_csv('/kaggle/input/titanic/train.csv')
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
data.drop(['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 1, inplace = True)
```

```python
test_data.drop(['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 1, inplace = True)
```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].mean(),inplace=True)

data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
test_data['Age'].fillna(test_data['Age'].mean(),inplace=True)
test_data['Fare'].fillna(test_data['Fare'].mean(),inplace=True)

test_data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sts = StandardScaler()
```

```python
feature_scale = ['Age', 'Fare']
data[feature_scale] = sts.fit_transform(data[feature_scale])
```

```python
test_data[feature_scale] = sts.fit_transform(test_data[feature_scale])
```

```python
data.head()
```

```python
test_data.head()
```

```python
women = data.loc[data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print("% of women who survived:", rate_women)
```

```python
men = data.loc[data.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men)
```

```python
X = pd.get_dummies(data.drop(['Survived'], axis = 1))
y = data['Survived']

X_test = pd.get_dummies(test_data)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('my_submission.csv', index=False)
```
